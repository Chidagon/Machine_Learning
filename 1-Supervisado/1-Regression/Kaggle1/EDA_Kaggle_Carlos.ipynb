```python
words_to_remove = ['Xiaomi','Apple','Samsung','Razer','Fujitsu','Microsoft','LG','Mediacom','Google','Vero','Huawei','Chuwi']
train = train[~train['Company'].isin(words_to_remove)]
train['Company'].value_counts()
```

```python
company_dict = {'Lenovo':1, 'Dell':2, 'HP':3, 'Asus':4, 'Acer':5, 'MSI':6, 'Toshiba':7}
train['Company'] = train['Company'].replace(company_dict)
train['Company']=train['Company'].astype('int64')
train['Company'].value_counts()
```

```python
palabras_to_remove = ['Workstation','Netbook']
train = train[~train['TypeName'].isin(palabras_to_remove)]
```

```python
TypeName_dict = {'Notebook':1, 'Ultrabook':2, 'Gaming':3, '2 in 1 Convertible':4}
train['TypeName'] = train['TypeName'].replace(TypeName_dict)
train['TypeName']=train['TypeName'].astype('int64')
train['TypeName'].value_counts()
```

```python
valores_to_remove = [12.5, 11.6, 13.9, 10.1, 18.4, 15.0, 14.1, 17.0]
train = train[~train['Inches'].isin(valores_to_remove)]
train['Inches'].value_counts()
```

```python

train['ScreenResolution'].value_counts()
```

```python
screen_to_remove = ['3840x2160', '3200x1800', '2560x1440', '1600x900','2304x1440','2256x1504','1440x900','2880x1800','2400x1600','2560x1600','2160x1440']
train = train[~train['ScreenResolution'].isin(screen_to_remove)]
train['ScreenResolution'].value_counts()
```

```python
Screen_dict = {'1920x1080':1, '1366x768':2}
train['ScreenResolution'] = train['ScreenResolution'].replace(Screen_dict)
```

```python
train["ScreenResolution"].astype('int64')
train["ScreenResolution"].value_counts()
```

```python
for index, row in train.iterrows():
    # Comprobar si la descripciÃ³n del disco duro contiene "SSD"
    if "SSD" in row["Memory"]:
        train.loc[index, "SSD"] = row["Memory"]
    # Comprobar si la descripciÃ³n del disco duro contiene "HDD"
    if "HDD" in row["Memory"]:
        train.loc[index, "HDD"] = row["Memory"]
```

```python
train["SSD"].value_counts()
```

```python
train['SSD']=train['SSD'].str[:5]
train['SSD']=train['SSD'].str.replace('+','')
train['SSD']=train['SSD'].str.replace('GB','')
train['SSD']=train['SSD'].str.replace('S','')
train['SSD']=train['SSD'].str.replace('TB','000')
#train['SSD']=train['SSD'].str.replace(' ','')
train['SSD']=train['SSD'].fillna(0)
train['SSD']
```

```python
train['SSD']=train['SSD'].astype("float64")
train['SSD']
```

```python
train['HDD']=train['HDD'].str[-10:]
train['HDD']=train['HDD'].str.replace('GB','')
train['HDD']=train['HDD'].str.replace('TB','000')
train['HDD']=train['HDD'].str.replace('+','')
train['HDD']=train['HDD'].str.replace(' ','')
train['HDD']=train['HDD'].str.replace('HDD','')
train['HDD']=train['HDD'].fillna(0)
train['HDD']

```

```python
train['HDD']=train['HDD'].astype("float64")
```
